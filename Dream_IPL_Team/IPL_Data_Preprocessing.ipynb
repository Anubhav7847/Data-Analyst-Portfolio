**2. Processing Batting Summary** 

**2.1 Reading the data**

```python
df_batting=pd.read_csv('C:/Users/SKS/Desktop/IPL project/batting_uncleaned.csv')
df_batting.head()
```

**2.2 Categorizing out/notout and renaming columns**

```python
df_batting['out/not_out']=df_batting['Unnamed: 1'].apply(lambda x: 'not_out' if x=='not out' else 'out')
df_batting['out/not_out'].value_counts()
```

```python
df_batting.drop(columns=['Unnamed: 1','M'], inplace=True)
```

```python
df_batting.rename({'R':'runs','B':'balls','BATTING':'batsmanName'}, axis=1, inplace=True)
df_batting.head()
```

**2.3 Cleanup weird characters**

```python
df_batting['batsmanName'].unique()
```

```python
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x: x.replace('â€',''))
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x: x.replace('\xa0',''))
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x: x.replace('†',''))
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x: x.replace('(c)',''))
```

```python
df_player=pd.read_csv('C:/Users/SKS/Desktop/IPL project/player_info.csv')
```

**2.4 Creating a teams dictionary that maps player name to country which they represent. This will be useful later on to link with other tables**

```python
player_dict=dict([(i,j) for i,j in zip(df_player.name,df_player.team)])
player_dict
```

```python
df_batting['team']=df_batting['batsmanName'].map(player_dict)
```

```python
df_batting.to_csv('C:/Users/SKS/Desktop/IPL project/batting_cleaned.csv',index=False)
```
